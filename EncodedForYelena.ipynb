```python
import random
import string

# 1. Generate substitution mapping (letters → unique ASCII chars)
def generate_substitution(seed=None):
    if seed is not None:
        random.seed(seed)  # Fix randomness for reproducibility

    letters = list(string.ascii_letters)  # 52 letters (A–Z, a–z)
    symbols = list(string.printable.strip())  # All printable ASCII (100+ chars)

    # Shuffle and assign unique symbols to letters
    random.shuffle(symbols)
    substitution_map = dict(zip(letters, symbols[:len(letters)]))

    return substitution_map

# 2. Encode text
def encode_text(text, substitution_map):
    return ''.join(substitution_map.get(ch, ch) for ch in text)

# 3. Decode text
def decode_text(encoded_text, substitution_map):
    reverse_map = {v: k for k, v in substitution_map.items()}
    return ''.join(reverse_map.get(ch, ch) for ch in encoded_text)


# Your long poem
text = """I love my sweet Armenia's word which is filled with the taste of sun,
I love our old lyre's melody from its mournful and weeping strings,
The vivacious fragrance of the blood-like flowers and the roses,
I love as well the graceful and agile dance of Nayirian girls.
I love as well our gloomy sky, our pure waters, luminous lake,
The summer's sun and the winter's sublime wind with a dragon's voice,
Also the black, unwelcoming walls of the huts lost in the dark,
And I love the thousand-year stone of the ancient cities as well.

No matter where I am yet I shall not forget our mournful songs,
Shall not forget our steel-lettered books which now have become prayers,
No matter how sharply they pierce my heart our wounds so soaked with blood,
Even then I love my orphaned and my bloodied, dear Armenia.

For my longing heart there is not, not even one another tale,
There's no brighter forehead than that of Kouchag and Naregatsi,
Pass the whole world, there's no summit as white as that of Ararat,
Like glory road, unreachable, I love as well my Mount Massis.
"""

# Generate substitution key
sub_map = generate_substitution(seed=42)

# Encode and decode
encoded = encode_text(text, sub_map)
decoded = decode_text(encoded, sub_map)

# Print results
print("Original:")
print(text, "\n")

print("Encoded:")
print(encoded, "\n")

print("Decoded:")
print(decoded, "\n")

# Verify correctness
print("Decoding matches original?", decoded == text)
```
